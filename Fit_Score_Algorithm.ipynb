```python
# (브라렛X) 앞중심 너비 설정 함수
def get_width_weight(product_width, user_width):
    if product_width == user_width:
        return 10
    elif user_width == 'E':
        return 5
    else: 
        return 0

def get_score(product, user, user_width):
    
    # 가중치 비중 - 제품 종류 : 컵 : 앞중심 높이 : 앞중심 너비 = 20 : 35 : 25 : 10
    product_type_weight = 25
    cup_weight = 35
    goreHeight_weight = 30
    goreWidth_weight = 10

    # 제품 종류 가중치: 35(추천O) - 0(추천X)
    if product['bra_type'] not in user['product_type']:
        product_type_weight = 0
    
    # 컵 가중치: 35(우선추천) - 17(일반추천) - 0(추천X)
    if product['cupCategory'] in user['cup_normal']:
        cup_weight = 17
    elif product['cupCategory'] not in user['cup_prior']:
        cup_weight = 0
    
    # 앞중심 높이 가중치: 35(우선추천) - 17(일반추천) - 0(추천X)
    if product['goreHeight'] in user['goreHeight_normal']:
        goreHeight_weight = 13
    elif product['goreHeight'] not in user['goreHeight_prior']:
        goreHeight_weight = 0
    
    # 앞중심 너비 가중치:
    # 브라렛) 10(추천O) - 5(추천X)
    # 브라렛X) 10(우선추천) - 5(일반추천) - 0(추천X) 
    if product['isBraret']:
        if product['goreWidth'] not in user['braretLine']:
            goreWidth_weight = 5
    elif len(user_type) == 3:
        goreWidth_weight = get_width_weight(product['goreWidth'], user_width)

    score = round((product_type_weight + cup_weight + goreHeight_weight + goreWidth_weight)/100, 2)
    if score == 1.0:
        score -= 0.03

    return score
```

# 추천도 추출 함수를 이용한 추천도 시트 출력

```python
import pandas as pd
import numpy as np

# 제품 csv 파일을 불러옵니다.
#bra_data = pd.read_csv("## 제품 csv 파일 경로 ##")

bra_data = ver4
# 제품 데이터 중 빈칸이 있는 행을 모두 제거합니다.
# bra_data = bra_data.dropna()
# bra_data = bra_data.reset_index(drop=True)

# 제품 종류 column을 추가합니다. 
# 'bra_type' : 'braret', 'wire', 'volume_up'
for i in bra_data.index:
    bra_type = []
    if bra_data.loc[i, 'isBraret']:
        bra_type.append('braret')
    if bra_data.loc[i, 'isNoWire'] == 0:
        bra_type.append('wire')
    else:
        bra_type.append('no_wire')
    if bra_data.loc[i, 'isVolumeUp'] == 1:
        bra_type.append('volume_up')
    if bra_data.loc[i, 'isBigSize'] == 1:
        bra_type.append('big')

    bra_data.loc[i, 'bra_type'] = ', '.join(bra_type)

bra_data['bra_type'] = bra_data['bra_type'].apply(lambda x: x.split(', '))

bra_data.info()
```

```python
import random

# 각 제품의 추천도를 저장할 data frame을 생성합니다.
data_score_df = pd.DataFrame(bra_data['Id'], columns=['Id'])

# criteria에 존재하는 가슴 너비를 제외한 가슴 유형에 대하여
# 각각 가슴 너비("C", "E", "W")를 덧붙이고 추천도를 계산합니다.
for user_type in criteria.keys():
    for user_width in ['C', 'E', 'W']:
        type_name = user_type + user_width

        # 알파벳 세 글자로 끝나는 유형(ex. medium_TEE)이 아닐 경우 type_name 표기를 다르게 합니다.
        if user_type in ['xsmall', 'small_bird', 'C_bird', 'large']:
            type_name = user_type + '_' + user_width

        # 모든 제품의 해당 유형에 대한 추천도를 계산하고 data_score_df에 추가합니다.
        for i in range(len(data_score_df)):
            product = bra_data.loc[i]
            score = get_score(product, criteria[user_type], user_width) + random.randrange(-9, 20)/100
            if score < 0:
                score = 0

            data_score_df.loc[i, type_name] = score
```

```python
def get_type_weight2(sample, type_1, type_2):
    type_weight = 40
    for x in sample['bra_type']:
        if x not in set(type_1['product_type']) & set(type_2['product_type']):
            type_weight = 0
    return type_weight

def get_cup_weight2(sample, type_1, type_2):
    cup_weight = 0
    if sample['cupCategory'] in set(type_1['cup_prior']) & set(type_2['cup_prior']):
        cup_weight = 20
    elif sample['cupCategory'] in set(type_1['cup_prior']) & set(type_2['cup_normal']) | set(type_1['cup_normal']) & set(type_2['cup_prior']):
        cup_weight = 13
    elif sample['cupCategory'] in set(type_1['cup_normal']) & set(type_2['cup_normal']):
        cup_weight = 6
    return cup_weight

def get_goreHeight_weight2(sample, type_1, type_2):
    goreHeight_weight = 0
    if sample['goreHeight'] in set(type_1['goreHeight_prior']) & set(type_2['goreHeight_prior']):
        goreHeight_weight = 20
    elif sample['goreHeight'] in set(type_1['goreHeight_prior']) & set(type_2['goreHeight_normal']) | set(type_1['goreHeight_normal']) & set(type_2['goreHeight_prior']):
        goreHeight_weight = 13
    elif sample['goreHeight'] in set(type_1['goreHeight_normal']) & set(type_2['goreHeight_normal']):
        goreHeight_weight = 6
    return goreHeight_weight

def get_braretWidth_weight2(sample, type_1, type_2):
    braretWidth_weight = 1
    if sample['goreWidth'] in set(type_1['braretWidth']) & set(type_2['braretWidth']):
        braretWidth_weight = 5
    elif sample['goreWidth'] in set(type_1['braretWidth']) | set(type_2['braretWidth']):
        braretWidth_weight = 3
    return braretWidth_weight
```

```python
# (브라렛X) 앞중심 너비 설정 함수
def get_width_weight(product_width, user_width):
    if product_width == user_width:
        return 5
    elif user_width == 'E':
        return 3
    else: 
        return 0

def get_score_urinnerf(product, user, user_width):
    # 가중치 초기화
    # 가중치 비중 - 컵 : 앞중심 높이 : 앞중심 너비 : 와이어유무 = 35 : 25 : 5 : 10
    cup_weight = 35
    goreHeight_weight = 25
    goreWidth_weight = 5
    wire_weight = 10
    
    # 컵 가중치: 35(우선추천) - 17(일반추천) - 0(추천X)
    if product['cupCategory'] in user['cup_normal']:
        cup_weight = 17
    elif product['cupCategory'] not in user['cup_prior']:
        cup_weight = 0
    
    # 앞중심 높이 가중치: 25(우선추천) - 14(일반추천) - 0(추천X)
    if product['goreHeight'] in user['goreHeight_normal']:
        goreHeight_weight = 14
    elif product['goreHeight'] not in user['goreHeight_prior']:
        goreHeight_weight = 0
    
    # 앞중심 너비 가중치: 5(우선추천) - 3(일반추천) - 0(추천X) 
    goreWidth_weight = get_width_weight(product['goreWidth'], user_width)

    # 와이어 유무 가중치: 10(추천) - 5(추천X)
    if user == 'large' and product['isWire'] == 0:
        wire_weight = 5

    score = round((cup_weight + goreHeight_weight + goreWidth_weight + wire_weight)/75, 2)
    if score == 1.0:
        score -= 0.03
    
    return score
```
